```
import matplotlib
import matplotlib.pyplot as plt
import os
import xml.etree.ElementTree as ET

ON_DRIVE = True
```

```
classes = ["car", "truck", "bicycle"]
n = len(classes)

currFolder = ""
if ON_DRIVE:
  currFolder = "drive/My Drive/Colab Notebooks/mqi-project/" 
trainFolder = currFolder + "data/train/annotation" 
testFolder = currFolder + "data/val/annotation"
```

```
numclasses = [0] * n
#for each file in annotation TRAINING
for filename in os.listdir(trainFolder):
  #read it and extract the classes
  tree = ET.parse(os.path.join(trainFolder,filename))
  root = tree.getroot()
  for annotation in root[6:]:  
    classname = annotation[0].text
    #count the classes
    for i in range(n):
      if(classname == classes[i]):
        numclasses[i] += 1
#plot classes in hystogram
plt.bar(classes, numclasses)
print(list(zip(classes, numclasses)))
```

```
numclasses = [0] * n
#for each file in annotation VALIDATION
for filename in os.listdir(testFolder):
  #read it and extract the classes
  tree = ET.parse(os.path.join(testFolder,filename))
  root = tree.getroot()
  for annotation in root[6:]:  
    classname = annotation[0].text
    #count the classes
    for i in range(n):
      if(classname == classes[i]):
        numclasses[i] += 1
#plot classes in hystogram
plt.bar(classes, numclasses)
print(list(zip(classes, numclasses)))
```

```
#this code computes the number of images that have the maximum number of items of each class
finalnumclasses = [0] * n
#for each file in annotation TRAINING
for filename in os.listdir(trainFolder):
  numclasses = [0] * n
  #read it and extract the classes
  tree = ET.parse(os.path.join(trainFolder,filename))
  root = tree.getroot()
  for annotation in root[6:]:  
    classname = annotation[0].text
    #count the classes
    for i in range(n):
      if(classname == classes[i]):
        numclasses[i] += 1
  ind = numclasses.index(max(numclasses))
  finalnumclasses[ind] += 1
#plot classes in hystogram
plt.bar(classes, finalnumclasses)
```
